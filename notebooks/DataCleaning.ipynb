```python
# Import libraries
import pandas as pd
```

```python
# Load the dataset
df = pd.read_csv('Employee-Attrition-and-Performance.csv')
```

```python
# Drop the garbage columns (LO5)
df_clean = df.drop(columns=['EmployeeCount', 'Over18', 'StandardHours', 'EmployeeNumber'], errors='ignore')
```

```python
# Rename columns to lowercase (Best practice)
df_clean.columns = [col.lower() for col in df_clean.columns]
```

```python
# Save the clean file
df_clean.to_csv('clean_employee_data.csv', index=False)

print("Clean file created.")
print("Original Shape:", df.shape)
print("New Shape:", df_clean.shape)
```
